### Automation 1: Pyautogui

In this jupyter we will review automation tools:

    - Pyautogui: As a package to create scripts that simulate mouse movements, clicks and hotkeys. 

```python
from time import sleep
import os
import pyautogui as pygui
import numpy as np
```

```python
Xactual, Yactual = pygui.position()
print (f"Current mouse position: {Xactual}, {Yactual}")

width, height = pygui.size()
print (f"Screen size: {width}x{height}")
```

```python
pygui.moveTo(Xactual, Yactual) 
pygui.click()

sleep(1)

pygui.moveTo(1660, 900)
pygui.click()
pygui.write('Im HERE!', interval=0.25)  
pygui.hotkey('ctrl', 'enter')
```

Drag a square on the screen

```python
def square(l):
    pygui.drag(l, 0, duration=1)
    pygui.drag(0, l, duration=1)
    pygui.drag(-l, 0, duration=1)
    pygui.drag(0, -l, duration=1)
```

```python
square(200)
```

```python
def open_text_editor():
    # Windows
    os.system("xed &")
    sleep(1)
    S = pygui.size()
    pygui.moveTo(S[0]//2, S[1]//2)
```

```python
sleep(2)
open_text_editor()
pygui.write('Im HERE!', interval=0.25)
```
