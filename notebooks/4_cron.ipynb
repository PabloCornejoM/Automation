### CRON

What is Cron? 

In Unix like OS, Cron is a process administrator that executes repetitive tasks in regular time intervals (minutes, days, weeks, months)  like daily backups, system maintenance, or report generation, configured in a special file called a crontab.



#### Basic Cron Commands

In terminal: <br>
**`crontab -l`** -> Shows list of program tasks<br>
**`crontab -e`** -> To edit program tasks
**`crontab -r`** -> To delete program tasks  <br>

Its important to manage tasks so they don't keep working indefinitely
